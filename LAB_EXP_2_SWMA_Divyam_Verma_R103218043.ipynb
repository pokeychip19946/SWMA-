#TARGET 2 : AMAZON SCRAPING

```
from bs4 import BeautifulSoup
import requests
```

STEP 1: SELECT A PRODUCT OF UR CHOICE WITH REVIEWS AND RATINGS 

HERE I HAD CHOSEN SAMSUNG GALAXY S10( PRISM BLUE, 8GB RAM,  128GB STORAGE)

```
link = "https://www.amazon.in/Samsung-Galaxy-Storage-Additional-Exchange/product-reviews/B07KXCK8S6/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
```

STEP 2: PULL REQUEST FOR REVIEWS AND RATING USING THE LINK OF THAT PRODUCT 

```
page = requests.get(link)
page
```

```
# USED TO PRINT THE WEBSITE SOURCE CODE
soup = BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

STEP 3: 

1. PRINT CUSTOMERS NAME WHO HAD REVIEWD AND RATED 
2. PRINT CUSTOMERS RATINGS RESPECTIVELY
3. PRINT CUSTOMER REVIEWS TITLE 
4. PRINT DATE OF REVIEW GIVEN BY A CUSTOMER
5. PRINT REVIEW'S TEXT PER CUSTOMER

NOTE: AMAZON RESTRICTS ONLY 10 -15 REVIEWS AND RATING TO BE SCRAPS 

```
names = soup.select('span.a-profile-name')[2:]
names
```

```
stars = soup.select('span.a-icon-alt')[3:]
stars
```

```
title = soup.select('a.review-title span')
title
```

```
date = soup.select('span.review-date')[2:]
date
```

```
data = soup.select('span.review-text span')
data
```

```
# LISTS ARE CREATED FOR CUSTOMER'S NAMES, RATINGS, REVIEW_DATE, REVIEW_TITLE AND REVIEW_TEXT TO APPEND ALL 10 DATA RESPECTEDLY  THAT ARE SCRAPED  
cust_name =[]
rating = []
review_date =[]
review_title=[]
review_text=[]
# LOOP TO ALL 10 DATAS COLLECTED IN THE RESPECTED LISTS IN TEXT FORM  
for i in range(10):
  cust_name.append(names[i].get_text())
  rating.append(stars[i].get_text())
  review_date.append(date[i].get_text().replace("Reviewed in India on ",""))# USED TO REPLACE TEXT FROM REVIEW DATE
  review_title.append(title[i].get_text())
  review_text.append(data[i].get_text())
```

STEP 4:   CREATE A DATAFRAME TO STORE LISTS DATA IN A TABULAR FORM OR STRUCTURED FORM 


```
df = pd.DataFrame()
df["Date"] = review_date
df["Customer Name"] = cust_name
df["Ratings"] = rating
df["Title"] = review_title
df["Review"] = review_text
df
```

STEP 5: STORE DATA IN A CSV LOCAL FILE

```
df.to_csv('reviews.csv')
```
